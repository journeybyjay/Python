### 1. urllib.parse.quote() : GET요청시 한글 URL 인코딩
### 2. urllib.request.urlopen() : HTTP 요청
### 3. response.getcode() : 응답코드 200 OK, 404이면 File Not Found
### 4. response.read() : html을 읽어옴

```python
import urllib.request

search_addr = "http://book.daum.net/search/bookSearch.do?query="
search_word = "파이썬"
search_word_enc = urllib.parse.quote(search_word)
print(search_word_enc)
URL = search_addr + search_word_enc
req = urllib.request.Request(URL)
response = urllib.request.urlopen(req)
print (response)
status_code = response.getcode()
print (status_code)
html_data = response.read()
print (html_data)
```

```python
import requests

URL = "http://book.daum.net/search/bookSearch.do1"
params = {'query' : '파이썬'}
headers = {'User-Agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 6_0 like Mac OS X) AppleWebKit/536.26 (KHTML, like Gecko) Version/6.0 Mobile/10A5376e Safari/8536.25'}

response = requests.get(URL, headers=headers, params=params)
print(dir(response))

status_code = response.status_code
if status_code == 200:
    html_data = response.text
    print(response.url)
    print(status_code)
else:
    print("Error Code is : " + str(status_code))
    print("Error is that reason :" + str(response.reason))
```

```python
import requests
from bs4 import BeautifulSoup as bs

params = {'query':'파이썬'}
url = "http://shopping.daum.net/search/o/"

res = requests.get(url, params = params)
html_content = bs(res.content, 'html.parser')
print(html_content.prettify())
```

```python
html_title = html_content.title
print(html_title.prettify())

html_body_span = html_content.body.span
print(html_body_span.prettify())

html_body_li = html_content.body.li
print(html_body_li.prettify())
```
